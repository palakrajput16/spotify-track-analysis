# Feature Engineering

In this notebook, we prepare the cleaned Spotify dataset for machine learning by creating new features and encoding categorical variables. The resulting dataset will be used for both popularity prediction and genre classification.


```python
import pandas as pd
import joblib

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("../data/processed/cleaned_dataset.csv")
```

```python
df.head()
```

## 5.1 Encode Track Genre

Machine learning models cannot work directly with text labels. Therefore, the `track_genre` column is converted into numerical labels using Label Encoding.

```python
le_genre = LabelEncoder()

df["genre_encoded"] = le_genre.fit_transform(df["track_genre"])

joblib.dump(le_genre, "../models/genre_label_encoder.pkl")
```

```python
df[["track_genre", "genre_encoded"]].head()
```

## 5.2 Convert Duration to Minutes

The original dataset stores song duration in milliseconds. Converting it to minutes makes the feature easier to interpret.

```python
df["duration_min"] = df["duration_ms"] / 60000
```

```python
df[["duration_ms", "duration_min"]].head()
```

## 5.3 Normalize Loudness

Spotify stores loudness in negative decibels. Min-max normalization converts it to a 0–1 scale, making it easier for machine learning models to process.

```python
df["loudness_norm"] = (
    df["loudness"] - df["loudness"].min()
) / (
    df["loudness"].max() - df["loudness"].min()
)
```

```python
df[["loudness", "loudness_norm"]].head()
```

## 5.4 Define Model Features

These audio features will be used as the input variables for both the popularity prediction and genre classification models.

```python
FEATURES = [
    "danceability",
    "energy",
    "key",
    "loudness",
    "mode",
    "speechiness",
    "acousticness",
    "instrumentalness",
    "liveness",
    "valence",
    "tempo",
    "duration_min",
    "time_signature",
    "explicit"
]
```

```python
FEATURES
```

## Save the Processed Dataset

The engineered dataset is saved for use in the modeling notebooks.

```python
df.to_csv("../data/processed/features_dataset.csv", index=False)
```
